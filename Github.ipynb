The Rössler attractor /ˈrɒslər/ is the attractor for the Rössler system, a system of three non-linear ordinary differential equations originally studied by Otto Rössler.[1][2] These differential equations define a continuous-time dynamical system that exhibits chaotic dynamics associated with the fractal properties of the attractor.[3] 

Attractor a set of numerical values toward which a system will naturally gravitate and remain cycling through unless perturbed [4]

__Variation of parameters__  <br>
Rössler attractor's behavior is largely a factor of the values of its constant parameters a, b, and c. In general, varying each parameter has a comparable effect by causing the system to converge toward a periodic orbit, fixed point, or escape towards infinity, however the specific ranges and behaviors induced vary substantially for each parameter. Periodic orbits, or "unit cycles," of the Rössler system are defined by the number of loops around the central point that occur before the loops series begins to repeat itself.[6]

```python
#import necessary libraries
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
from pylab import figure, show, setp
from mpl_toolkits.mplot3d import Axes3D

# set the parameters
a = 0.3
b = 0.3
c = 5.99

# create the function
def rossler_attractor(X,t):
    x, y, z = X    
    dxdt = -y - z
    dydt = x + a*y
    dzdt = b*x - c*z + x*z
    return [dxdt, dydt, dzdt]

# numerical integration
X0 = [1, 1, 1]
t = np.linspace(0, 200, 100000)
result = odeint(rossler_attractor, X0, t)
x, y, z = result.T

# create figures 
fig = plt.figure(figsize=(13,9))
ax1 = fig.add_axes([0.1, 0.7, 0.4, 0.2])
ax2 = fig.add_axes([0.1, 0.4, 0.4, 0.2])
ax3 = fig.add_axes([0.1, 0.1, 0.4, 0.2])
ax4 = fig.add_axes([0.55, 0.1, 0.8, 0.8],projection='3d')

# plot of x and t
ax1.plot(t, x,color='red',lw=1,label='x(t)')
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')
ax1.legend()
ax1.axis((min(t),max(t),min(x),max(x)))

# plot of y and t
ax2.plot(t, y,color='green',lw=1,label='y(t)')
ax2.set_xlabel('t')
ax2.set_ylabel('y(t)')
ax2.legend()
ax2.axis((min(t),max(t),min(y),max(y)))

# plot of z and t
ax3.plot(t, z,color='blue',lw=1,label='z(t)')
ax3.set_xlabel('t')
ax3.set_ylabel('z(t)')
ax3.legend()
ax3.axis((min(t),max(t),min(z),max(z)))

# plot of the rossler system
ax4.plot(x, y,z,color='magenta',lw=1,label='Evolution(t)')
ax4.set_xlabel('x(t)')
ax4.set_ylabel('y(t)')
ax4.set_zlabel('z(t)')
ax4.set_title('Rossler System')
ax4.view_init(20, 160)

plt.show()
```

__Varying a__  <br>
Here, b is fixed at 0.2, c is fixed at 5.7 and a changes. Numerical examination of the attractor's behavior over changing a suggests it has a disproportional influence over the attractor's behavior. The results of the analysis are: <br>
<br>
a≤0:
<br>
a≤0: converges to the centrally located fixed point.
<br>
a = 0.1: unit cycle of period 1.
<br>
a = 0.2: standard parameter value selected by Rössler, chaotic.
<br>
a = 0.3: chaotic attractor, significantly more Möbius strip-like (folding over itself).
<br>
a = 0.35: similar to 0.3, but increasingly chaotic.
<br>
a = 0.38: Similar to 0.35, but increasingly chaotic. [7]
<br>

```python
 #Varying a. b and c are fixed
a = [0.1, 0.1, 0.2, 0.3, 0.35, 0.38, 0.45]
b = 0.2
c = 5.7

def rossler_attractor(X,t):
    x, y, z = X    
    dxdt = -y - z
    dydt = x + a[aiter]*y
    dzdt = dzdt = b*x - c*z + x*z
    return [dxdt, dydt, dzdt]

for aiter in range(len(a)):
    X0 = [1, 1, 1]
    t = np.linspace(0, 200, 100000)
    result = odeint(rossler_attractor, X0, t)
    x, y, z = result.T

    fig = plt.figure(figsize=(8,5))
    ax4 = fig.add_axes([0.55, 0.1, 0.8, 0.8],projection='3d')

    ax4.plot(x,y,z,color='magenta',lw=1,label='Evolution(t)')
    ax4.set_xlabel('x(t)')
    ax4.set_ylabel('y(t)')
    ax4.set_zlabel('z(t)')
    ax4.set_title('Rossler System, a = %f' %a[aiter])
    ax4.view_init(20, 160)

    plt.show()
```

__Varying b__ <br>
Here, a is fixed at 0.2, c is fixed at 5.7 and b changes. As shown in the accompanying diagram, as b approaches 0 the attractor approaches infinity (note the upswing for very small values of b. Comparative to the other parameters, varying b generates a greater range when period-3 and period-6 orbits will occur. In contrast to a and c, higher values of b converge to period-1, not to a chaotic state. [8]

```python
#varying b. a and c are fixed 
a = 0.2
b = [0.2, 0.5, 0.8, 1.1, 1.4, 2, 5, 10, 15, 30, 50]
c = 5.7

def rossler_attractor(X,t):
    x, y, z = X    
    dxdt = -y - z
    dydt = x + a*y
    dzdt = b[biter]*x - c*z + x*z
    return [dxdt, dydt, dzdt]

for biter in range(len(b)):
    X0 = [1, 1, 1]
    t = np.linspace(0, 200, 100000)
    result = odeint(rossler_attractor, X0, t)
    x, y, z = result.T

    fig = plt.figure(figsize=(8,5))
    ax4 = fig.add_axes([0.55, 0.1, 0.8, 0.8],projection='3d')

    ax4.plot(x,y,z,color='magenta',lw=1,label='Evolution(t)')
    ax4.set_xlabel('x(t)')
    ax4.set_ylabel('y(t)')
    ax4.set_zlabel('z(t)')
    ax4.set_title('Rossler System, b = %f' %b[biter])
    ax4.view_init(20, 160)

    plt.show()
```

__Varying c__  <br>
Here, a = b = 0.1 and c changes. The bifurcation diagram reveals that low values of c are periodic, but quickly become chaotic as c increases. This pattern repeats itself as c increases --- there are sections of periodicity interspersed with periods of chaos, and the trend is towards higher-period orbits as c increases. For example, the period one orbit only appears for values of c around 4 and is never found again in the bifurcation diagram. The same phenomenon is seen with period three; until c = 12, period three orbits can be found, but thereafter, they do not appear. <br>
<br>
c = 4, period-1 orbit.
<br>
c = 6, period-2 orbit.
<br>
c = 8.5, period-4 orbit.
<br>
c = 8,7, period-8 orbit.
<br>
c = 9,sparse chaotic attractor.
<br>
c = 12, period-3 orbit.
<br>
c = 12.6, period-6 orbit
<br>
c = 13, sparse chaotic attractor.
<br>
c = 18, filled-in chaotic attractor. [9]

```python
#Varying c. a and c are fixed
a = 0.1
b = 0.1
c = [4, 6, 8.5, 8.7, 9, 12, 12.6, 13, 18]

def rossler_attractor(X,t):
    x, y, z = X    
    dxdt = -y - z
    dydt = x + a*y
    dzdt = dzdt = b*x - c[citer]*z + x*z
    return [dxdt, dydt, dzdt]

for citer in range(len(c)):
    X0 = [1, 1, 1]
    t = np.linspace(0, 200, 100000)
    result = odeint(rossler_attractor, X0, t)
    x, y, z = result.T

    fig = plt.figure(figsize=(8,5))
    ax4 = fig.add_axes([0.55, 0.1, 0.8, 0.8],projection='3d')

    ax4.plot(x,y,z,color='magenta',lw=1,label='Evolution(t)')
    ax4.set_xlabel('x(t)')
    ax4.set_ylabel('y(t)')
    ax4.set_zlabel('z(t)')
    ax4.set_title('Rossler System, c = %f' %c[citer])
    ax4.view_init(20, 160)

    plt.show()
```

__Insights:__

- When a increases from 0.2 , the system becomes increasingly becomes chaotic.


- When b approaches 0, the attractor approaches infinity. Meanwhile, higher values of b converges to a not chaotic state.


- When c is up to 4, he system becomes periodic while if it’s more than 5.7, the system becomes chaotic.
