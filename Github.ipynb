The Rössler attractor /ˈrɒslər/ is the attractor for the Rössler system, a system of three non-linear ordinary differential equations originally studied by Otto Rössler.[1][2] These differential equations define a continuous-time dynamical system that exhibits chaotic dynamics associated with the fractal properties of the attractor.[3] 

Attractor a set of numerical values toward which a system will naturally gravitate and remain cycling through unless perturbed

ROSSLER ATTRACTOR the attractor for Rossler system, a system of non-linear ordinary differential equations originally studied by Otto Rossler


$$
Rossler Attractor
$$

$$
\frac{dx}{dt} = -y-z
$$

$$
\frac{dy}{dt} = x + ay
$$

$$
\frac{dz}{dt} = b + z(x-c)
$$

```python
#import necessary libraries
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
from pylab import figure, show, setp
from mpl_toolkits.mplot3d import Axes3D

# set the parameters
a = 0.3
b = 0.3
c = 5.99

# create the function
def rossler_attractor(X,t):
    x, y, z = X    
    dxdt = -y - z
    dydt = x + a*y
    dzdt = b*x - c*z + x*z
    return [dxdt, dydt, dzdt]

# numerical integration
X0 = [1, 1, 1]
t = np.linspace(0, 200, 100000)
result = odeint(rossler_attractor, X0, t)
x, y, z = result.T

# create figures 
fig = plt.figure(figsize=(13,9))
ax1 = fig.add_axes([0.1, 0.7, 0.4, 0.2])
ax2 = fig.add_axes([0.1, 0.4, 0.4, 0.2])
ax3 = fig.add_axes([0.1, 0.1, 0.4, 0.2])
ax4 = fig.add_axes([0.55, 0.1, 0.8, 0.8],projection='3d')

# plot of x and t
ax1.plot(t, x,color='red',lw=1,label='x(t)')
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')
ax1.legend()
ax1.axis((min(t),max(t),min(x),max(x)))

# plot of y and t
ax2.plot(t, y,color='green',lw=1,label='y(t)')
ax2.set_xlabel('t')
ax2.set_ylabel('y(t)')
ax2.legend()
ax2.axis((min(t),max(t),min(y),max(y)))

# plot of z and t
ax3.plot(t, z,color='blue',lw=1,label='z(t)')
ax3.set_xlabel('t')
ax3.set_ylabel('z(t)')
ax3.legend()
ax3.axis((min(t),max(t),min(z),max(z)))

# plot of the rossler system
ax4.plot(x, y,z,color='magenta',lw=1,label='Evolution(t)')
ax4.set_xlabel('x(t)')
ax4.set_ylabel('y(t)')
ax4.set_zlabel('z(t)')
ax4.set_title('Rossler System')
ax4.view_init(20, 160)

plt.show()
```

```python
#Varying c. a and c are fixed
a = 0.1
b = 0.1
c = [4, 6, 8.5, 8.7, 9, 12, 12.6, 13, 18]

def rossler_attractor(X,t):
    x, y, z = X    
    dxdt = -y - z
    dydt = x + a*y
    dzdt = dzdt = b*x - c[citer]*z + x*z
    return [dxdt, dydt, dzdt]

for citer in range(len(c)):
    X0 = [1, 1, 1]
    t = np.linspace(0, 200, 100000)
    result = odeint(rossler_attractor, X0, t)
    x, y, z = result.T

    fig = plt.figure(figsize=(8,5))
    ax4 = fig.add_axes([0.55, 0.1, 0.8, 0.8],projection='3d')

    ax4.plot(x,y,z,color='magenta',lw=1,label='Evolution(t)')
    ax4.set_xlabel('x(t)')
    ax4.set_ylabel('y(t)')
    ax4.set_zlabel('z(t)')
    ax4.set_title('Rossler System, c = %f' %c[citer])
    ax4.view_init(20, 160)

    plt.show()
```

__Insights:__

- When a increases from 0.2 , the system becomes increasingly becomes chaotic.


- When b approaches 0, the attractor approaches infinity. Meanwhile, higher values of b converges to a not chaotic state.


- When c is up to 4, he system becomes periodic while if it’s more than 5.7, the system becomes chaotic.
